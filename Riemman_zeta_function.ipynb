# Riemman Zeta Function

The Riemman Zeta function is define as following: $$\zeta(s)=\sum_{n=1}^{+\infty}\dfrac{1}{n^s}$$ with $s$ in the Complex numbers.

```python
import mpmath
import numpy as np
import matplotlib.pyplot as plt
```

```python
def zeta_riemann(s):
    return mpmath.zeta(s)
```

```python
real_part = 0.5  # Fijar la parte real
imag_vals = np.linspace(-20, 20, 400)  # Valores para la parte imaginaria
zeta_values = [zeta_riemann(complex(real_part, y)) for y in imag_vals]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(imag_vals, [abs(z) for z in zeta_values], label='|(0.5 + yi)|')
plt.title('Valor Absoluto de la Funcin Zeta de Riemann a lo largo de la lnea Re(s) = 0.5')
plt.xlabel('Parte Imaginaria de s (yi)')
plt.ylabel('|(s)|')
plt.legend()
plt.grid(True)
plt.show()
```
