# Diferenciacion Numerica

Vamos a utilizar diferencias finitas para aproximar derivadas.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline
```

```python
# tamano de paso
h = 0.1

# hacemos una malla
x = np.arange(0, 2*np.pi, h) 

# calculemos una funcion sobre la malla
y = np.cos(x) 

# calculemos un vector de diferencias finitas hacia adelante
forward_diff = np.diff(y)/h 

# calculemos la malla correspondiente
x_diff = x[:-1:] 

# calculemos la solucion exacta
exact_solution = -np.sin(x_diff) 

# graficamos la solucion
plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', \
         label = 'Aproximacion de Diferencias Finitas')
plt.plot(x_diff, exact_solution, \
         label = 'Solucion exacta')
plt.legend()
plt.show()

# Compute max error between 
# numerical derivative and exact solution
max_error = max(abs(exact_solution - forward_diff))
print(max_error)
```

```python
# tamano de paso
h = 1

# numero de iteraciones
iterations = 20

# lsita para guardar los tamanos de los pasos
step_size = []

# lista para guardar el error maximo de cada paso
max_error = []

for i in range(iterations):
    # reducimos el tamano del paso
    h /= 2

    # store this step size
    step_size.append(h)

    # calculamos la nueva malla
    x = np.arange(0, 2 * np.pi, h)

    # calculamos el valor de la funcion sobre la malla
    y = np.cos(x)

    # calculamos el vector de diferencias finitas hacia adelante
    forward_diff = np.diff(y)/h

    # calculamos la malla correspondiente
    x_diff = x[:-1]

    # calculamos la solucion exacta
    exact_solution = -np.sin(x_diff)
    
    # calculamos el error maximo entre la derivada numerica y la solucion exacta\
    max_error.append(\
            max(abs(exact_solution - forward_diff)))

# producimos n grafico log-log del error maximo vs el tamano de paso
plt.figure(figsize = (12, 8))
plt.loglog(step_size, max_error, 'v')
plt.show()
```

## Diferenciacion numerica con ruido

```python
x = np.arange(0, 2*np.pi, 0.01) 
# calculamos la funcion
omega = 100
epsilon = 0.01

y = np.cos(x) 
y_noise = y + epsilon*np.sin(omega*x)

# graficamos la solucion
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = 'cos(x) + noise')
plt.plot(x, y, 'b-', \
         label = 'cos(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.01) 
# calculamos la funcion
y = -np.sin(x) 
y_noise = y + epsilon*omega*np.cos(omega*x)

# graficamos la solucion
plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = 'Derivada de cos(x) + noise')
plt.plot(x, y, 'b-', \
         label = 'Derivada de cos(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```
